W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input())
demchan = 0
demle = 0
if n == 0:
    demchan = 1
    demle = 0
else:
    n = abs(n)
    while n > 0:
        if n % 2 == 0:
            n = n // 10
            demchan = demchan + 1
        else:
            n = n // 10
            demle = demle + 1
print(f'số chữ số chẵn là {demchan} \n số chữ số lẻ là {demle}')
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập


```python
n = int(input())
max = n
min = n
while True:
    n = int(input())
    if n == -1:
        break  
    if n > max:
        max = n
    if n < min:
        min = n
print(max)
print(min)
```

W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
A = int(input())
a, b = 1, 1
while b <= A:
    a, b = b, a + b
print(a)
```

W4A27
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ v

```python
s = input()

count = 0
in_word = False

for ch in s:
    if ch != ' ' and not in_word:
        count += 1
        in_word = True
    elif ch == ' ':
        in_word = False

print(count)
```

W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành
chữ thường và ngược lại


```python
s = input()

result = ""

for ch in s:
    if 'a' <= ch <= 'z':
        result += chr(ord(ch) - 32)
    elif 'A' <= ch <= 'Z':
        result += chr(ord(ch) + 32)
    else:
        result += ch

print(result)
```

W4A30
Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính
tổng 3 số nguyên đó
VD:
Nhập: 3, 12, 15
Tổng: 30

```python
s = input()

num = ""
total = 0

for ch in s:
    if ch == ',':
        total += int(num)
        num = ""
    elif ch != ' ':
        num += ch

total += int(num)

print(total)
```

W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nh

```python
s = input()

upper = 0
lower = 0

for ch in s:
    if 'A' <= ch <= 'Z':
        upper += 1
    elif 'a' <= ch <= 'z': 
        lower += 1

print( upper)
print( lower)
```
